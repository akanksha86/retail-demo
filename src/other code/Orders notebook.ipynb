```python
# @title Setup bigquery client and formatting
from google.cloud import bigquery
from google.colab import data_table

project = 'esdemo-389207' # Project ID
client = bigquery.Client(project=project)
data_table.enable_dataframe_formatter()
# @title Executes the query, using client from previous cell
sql_top_10_clothing_products_by_total_sales = '''# Get the top 10 products in the clothing category by total sales

SELECT
  products.name,
  SUM(order_items.total_price) AS total_sales
FROM
  `esdemo-389207`.`retail_data`.`order_items` AS order_items
INNER JOIN
  `esdemo-389207`.`retail_data`.`products` AS products
ON
  order_items.product_id = products.product_id
WHERE
  products.category = 'Clothing'
GROUP BY
  products.name
ORDER BY
  total_sales DESC
LIMIT
  10;'''
query_top_10_clothing_products_by_total_sales = client.query(sql_top_10_clothing_products_by_total_sales)

# store the destination to reference in future cells
query_destination_top_10_clothing_products_by_total_sales = query_top_10_clothing_products_by_total_sales.destination.project + '.' + query_top_10_clothing_products_by_total_sales.destination.dataset_id + '.' + query_top_10_clothing_products_by_total_sales.destination.table_id

#  Render the query results
job_top_10_clothing_products_by_total_sales= client.get_job(query_top_10_clothing_products_by_total_sales.job_id)
df_top_10_clothing_products_by_total_sales = job_top_10_clothing_products_by_total_sales.to_dataframe()
df_top_10_clothing_products_by_total_sales
```

```python

# @title Executes the query, using client from previous cell
sql_top_10_products_2025 = '''# Fetch the top 10 products sold during 2025, including product name and quantity sold

SELECT
  products.name,
  SUM(order_items.quantity) AS total_quantity_sold
FROM
  `esdemo-389207`.`retail_data`.`order_items` AS order_items
INNER JOIN
  `esdemo-389207`.`retail_data`.`products` AS products
ON
  order_items.product_id = products.product_id
INNER JOIN
  `esdemo-389207`.`retail_data`.`orders_dp` AS orders
ON
  order_items.order_id = orders.order_id
WHERE
  DATE(orders.order_date) BETWEEN '2025-01-01'
  AND '2025-12-31'
GROUP BY
  products.name
ORDER BY
  total_quantity_sold DESC
LIMIT
  10;'''
query_top_10_products_2025 = client.query(sql_top_10_products_2025)

# store the destination to reference in future cells
query_destination_top_10_products_2025 = query_top_10_products_2025.destination.project + '.' + query_top_10_products_2025.destination.dataset_id + '.' + query_top_10_products_2025.destination.table_id

#  Render the query results
job_top_10_products_2025= client.get_job(query_top_10_products_2025.job_id)
df_top_10_products_2025 = job_top_10_products_2025.to_dataframe()
df_top_10_products_2025
```
